## SMART STUDENT MANAGEMENT SYSTEM


```python
FILE = "students.txt"
open(FILE, "a").close()

def add_student():
    print("\nAdd Student")
    while True:
        sid = input("\nEnter Student ID (integer): ")
        if sid.isdigit():
            break
        print("Invalid. ID must be a number.")
    with open(FILE, "r") as f:
        for line in f:
            if line.startswith(sid + ","):
                print("ID already exists.")
                return
    while True:
        name = input("Enter Name: ").strip()
        if not name or name.isnumeric():
            print("Invalid Name.")
        else:
            break
    while True:
        age = input("Enter Age: ")
        if age.isdigit() and int(age) > 0:
            break
        print("Invalid Age.")
    while True:
        grade = input("Enter Grade (A-F): ").upper()
        if grade in ["A","B","C","D","E","F"]:
            break
        else:
            print("Invalid Grade.")
    while True:
        marks = input("Enter Marks (0-100): ")
        if marks.isdigit() and 0 <= int(marks) <= 100:
            break
        print("Invalid Marks.")
    with open(FILE, "a") as f:
        f.write(f"{sid},{name},{age},{grade},{marks}\n")
    print("Student Added Successfully.")

def view_students():
    print("\nAll Students")
    count = 0
    with open(FILE, "r") as f:
        for line in f:
            print(line.strip())
            count += 1
    print("Total Records:", count)

def search_student():
    print("\nSearch Student")
    key = input("Enter ID or Name to search: ").lower()
    found = False
    with open(FILE, "r") as f:
        for line in f:
            if key in line.lower():
                print(line.strip())
                found = True
    if not found:
        print("No record found.")

def update_student():
    print("\nUpdate Student")
    sid = input("Enter Student ID to update: ")
    lines = []
    updated = False
    with open(FILE, "r") as f:
        lines = f.readlines()
    with open(FILE, "w") as f:
        for line in lines:
            data = line.strip().split(",")
            if data[0] == sid:
                print("Record Found:", line.strip())
                name = input("Enter New Name: ")
                age = input("Enter New Age: ")
                grade = input("Enter New Grade: ").upper()
                marks = input("Enter New Marks: ")
                f.write(f"{sid},{name},{age},{grade},{marks}\n")
                updated = True
            else:
                f.write(line)
    if updated:
        print("Record Updated.")
    else:
        print("ID Not Found.")

def delete_student():
    print("\nDelete Student")
    sid = input("Enter Student ID to delete: ")
    lines = []
    deleted = False
    with open(FILE, "r") as f:
        lines = f.readlines()
    with open(FILE, "w") as f:
        for line in lines:
            if not line.startswith(sid + ","):
                f.write(line)
            else:
                deleted = True
    if deleted:
        print("Record Deleted.")
    else:
        print("ID Not Found.")

def analyze_data():
    print("\nData Analysis")
    marks_list = []
    names = []
    with open(FILE, "r") as f:
        for line in f:
            data = line.strip().split(",")
            names.append(data[1])
            marks_list.append(int(data[4]))
    if not marks_list:
        print("No data available.")
        return
    avg = sum(marks_list) / len(marks_list)
    highest = max(marks_list)
    lowest = min(marks_list)
    top_index = marks_list.index(highest)
    top_name = names[top_index]
    below_avg = sum(1 for m in marks_list if m < avg)
    print("Average Marks:", round(avg, 2))
    print("Top Performer:", top_name, "(", highest, ")")
    print("Students Below Average:", below_avg)
    print("Highest Marks:", highest, "| Lowest Marks:", lowest)

def main():
    while True:
        print("\nSmart Student Record Analyzer")
        print("1. Add Student")
        print("2. View All Students")
        print("3. Search Student")
        print("4. Update Student")
        print("5. Delete Student")
        print("6. Analyze Data")
        print("7. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            add_student()
        elif choice == "2":
            view_students()
        elif choice == "3":
            search_student()
        elif choice == "4":
            update_student()
        elif choice == "5":
            delete_student()
        elif choice == "6":
            analyze_data()
        elif choice == "7":
            print("Exiting Program.")
            break
        else:
            print("Invalid Option.")
```
